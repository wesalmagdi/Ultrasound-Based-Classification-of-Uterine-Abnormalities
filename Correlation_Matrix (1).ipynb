```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/Ultrasound-Based-Classification-of-Uterine-Abnormalities/data/labels.csv')
df.head()
```

```python
df['label'] = df['label'].map({
    'normal': 0,
    'abnormal': 1
})
```

```python
import pandas as pd

# Assuming your dataframe is called df
# And your target column is called "target"
# (replace with your actual column name)

target = "label"   # <-- change this

# 1. Calculate correlation of every feature with the target
target_correlations = df.corr()[target].drop(target)

# 2. Sort by strongest correlation
target_correlations = target_correlations.abs().sort_values(ascending=False)

print("Correlation of each feature with the target:")
print(target_correlations)
```

```python
threshold = 0.02

important_features = target_correlations[target_correlations > threshold].index.tolist()

df_selected = df[important_features + ['label']]

print("Number of features kept:", len(important_features))
print("Features kept:", important_features)
```

```python
duplicates = df.T.duplicated()
duplicate_columns = df.columns[duplicates]

print("Duplicate features:")
print(duplicate_columns)
```

```python
df_selected = df_selected.drop(columns=duplicate_columns, errors='ignore')

print("Number of features after dropping duplicates:", df_selected.shape[1] - 1) # Subtract 1 for the 'label' column
print("Remaining features:", df_selected.columns.tolist())
```
